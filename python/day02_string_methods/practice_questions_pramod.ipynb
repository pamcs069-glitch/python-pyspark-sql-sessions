## Q1 — 🟢 Easy
### CSV Row Parser

---

### Concept Warm-Up — split, strip, dict(zip(...))

````python
# Warm-up 1: split on comma — spaces are NOT removed automatically
row = "Alice Johnson, Engineering, 95000, Mumbai"

raw_parts   = row.split(",")
clean_parts = [p.strip() for p in row.split(",")]

print(raw_parts)    # spaces survive in 'raw_parts'
print(clean_parts)  # clean_parts strips each element
````

````python
# Warm-up 2: dict(zip(headers, values)) — build a record from two lists
headers = ["name", "dept", "salary"]
values  = ["Alice Johnson", "Engineering", "95000"]

record = dict(zip(headers, values))
print(record)
````

````python
# Warm-up 3: combine split + strip + zip — the full parse pattern
header_line = "name, dept, salary, city"
data_line   = "Alice Johnson,  Engineering ,95000,  Mumbai  "

headers = [h.strip() for h in header_line.split(",")]
values  = [v.strip() for v in data_line.split(",")]
record  = dict(zip(headers, values))
print(record)
````

### Problem Statement

You receive a batch of CSV rows where the **first row** is the header and the rest are data rows.  
Some rows have extra whitespace around values — these must be stripped.

Write a function `parse_csv_batch(lines: list) -> list` that:
1. Treats the first element as the header row
2. Parses each subsequent row into a dict using `split(",")` + `strip()` + `dict(zip(...))`
3. Returns a list of dicts

---

### Input

```python
lines = [
    "customer_id, full_name, email, city",
    "101,  Alice Johnson , alice@gmail.com , Mumbai  ",
    "102, Bob Smith,bob@yahoo.com,Delhi",
    "103,  CAROL WHITE , carol@hotmail.com ,  Chennai  ",

    "105,eve nair  ,EVE@gmail.com,  Bangalore",
```

### Expected Output

```python
[
    {'customer_id': '101', 'full_name': 'Alice Johnson', 'email': 'alice@gmail.com', 'city': 'Mumbai'},
    {'customer_id': '102', 'full_name': 'Bob Smith',     'email': 'bob@yahoo.com',   'city': 'Delhi'},
    {'customer_id': '103', 'full_name': 'CAROL WHITE',   'email': 'carol@hotmail.com', 'city': 'Chennai'},
    {'customer_id': '104', 'full_name': 'David Brown',   'email': 'david@gmail.com', 'city': 'Pune'},
    {'customer_id': '105', 'full_name': 'eve nair',      'email': 'EVE@gmail.com',   'city': 'Bangalore'},
]
```

---

### Constraints
- Must use `split(",")` — no `csv` module
- Must strip each field with `strip()`
- Must use `dict(zip(...))` to build each record
- Do NOT normalise case — return values exactly as they appear after strip

````python
# Q1 — Your solution here
lines = [
    "customer_id, full_name, email, city",
    "101,  Alice Johnson , alice@gmail.com , Mumbai  ",
    "102, Bob Smith,bob@yahoo.com,Delhi",
    "103,  CAROL WHITE , carol@hotmail.com ,  Chennai  ",
    "104, David Brown,david@gmail.com,Pune",
    "105,eve nair  ,EVE@gmail.com,  Bangalore",
]

def parse_csv_batch(lines: list) -> list:
    header=lines[0].split(',')
    final_output=[]
    for line in lines[1:]:
        values =[v.strip() for v in line.split(',')]
        final_output.append(dict(zip(header,values)))
    return final_output
        


records = parse_csv_batch(lines)
for r in records:
    print(r)
````

### Email Domain Analyser

---

### Concept Warm-Up — split('@'), count(), startswith, isdigit()

````python
# Warm-up 1: extract domain from email using split('@')
email = "alice@gmail.com"

parts  = email.split("@")
local  = parts[0]   # 'alice'
domain = parts[1]   # 'gmail.com'

print(f"domain = {domain}")
print(f"local  = {local}")
````

````python
# Warm-up 2: validate email has exactly one '@' and contains '.'
emails = ["alice@gmail.com", "bob AT yahoo DOT com", "carol@hotmail.com"]

for email in emails:
    is_valid = email.count("@") == 1 and "." in email.split("@")[1]
    print(is_valid)
````

````python
# Warm-up 3: accumulate counts into a dict using dict.get(key, default)
emails = [
    "alice@gmail.com", "bob@yahoo.com", "carol@gmail.com",
    "dave@hotmail.com", "eve@gmail.com", "frank@yahoo.com",
]

counts = {}
for email in emails:
    domain = email.split("@")[1].lower().strip()
    counts[domain] = counts.get(domain, 0) + 1

print(counts)
````

````python
# Warm-up 4: sort dict by value descending
counts = {'gmail.com': 3, 'yahoo.com': 2, 'hotmail.com': 1}
sorted_counts = sorted(counts.items(), key=lambda x: -x[1])
print(sorted_counts)
````

---

### Concept Warm-Up — key=value parsing, isdigit(), replace chains, f-strings

````python
# Warm-up 1: parse key=value tokens from a list of parts
parts = ["user=john", "action=login", "status=failed"]

kv = {}
for part in parts:
    if "=" in part:
        key, value = part.split("=", 1)   # maxsplit=1 handles value containing '='
        kv[key] = value

print(kv)
````

````python
# Warm-up 2: clean phone using isdigit() filter and replace chains
phones = ["+91-98765-43210", "(987) 654-3210", "987.654.3210"]

for phone in phones:
    clean = "".join(c for c in phone if c.isdigit())
    print(clean)
````

````python
# Warm-up 3: multi-condition validation with if/elif
test_emails = ["alice@gmail.com", "bob@yahoo.com", "carol_at_gmail", "dave@gmail"]

for email in test_emails:
    email = email.strip().lower()
    if email.count("@") != 1:
        print("invalid — no @")
    elif "." not in email.split("@")[1]:
        print("invalid — no domain dot")
    else:
        print("valid")
````
