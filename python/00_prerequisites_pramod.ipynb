## Step 5 — Quick Smoke Test

Test that standard library modules work correctly — this is what Day notebooks use.

````python
# csv module
import csv, io
sample = "id,name,salary\n1,Alice,95000\n2,Bob,65000"
reader = csv.DictReader(io.StringIO(sample))
rows = list(reader)
assert len(rows) == 2
print(f"csv module    — OK ({len(rows)} rows read)")

# json module
import json
data = json.loads('{"name": "Alice", "salary": 95000}')
assert data["name"] == "Alice"
print(f"json module   — OK")

# sqlite3
import sqlite3
conn = sqlite3.connect(":memory:")
conn.execute("CREATE TABLE t (id INT, name TEXT)")
conn.execute("INSERT INTO t VALUES (1, 'Alice')")
result = conn.execute("SELECT * FROM t").fetchall()
assert result == [(1, 'Alice')]
conn.close()
print(f"sqlite3       — OK")

# collections
from collections import defaultdict, Counter, deque
c = Counter([1,1,2,3,3,3])
assert c[3] == 3
print(f"collections   — OK")

# heapq
import heapq
top3 = heapq.nlargest(3, [5, 1, 9, 3, 7])
assert top3 == [9, 7, 5]
print(f"heapq         — OK")

# itertools
import itertools
groups = [(k, list(g)) for k, g in itertools.groupby([1,1,2,2,3], key=lambda x: x)]
assert len(groups) == 3
print(f"itertools     — OK")

print("\nAll checks passed. You are ready to start Day 1.")
````
