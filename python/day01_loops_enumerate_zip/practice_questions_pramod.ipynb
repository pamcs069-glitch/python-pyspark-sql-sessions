## Q1 — 🟢 Easy
### Inventory Status Report

---

### Problem Statement

You are given two lists of equal length:
- `products` — product names
- `stock` — current stock quantity for each product

Using **`enumerate`**, loop over both and build an inventory report as a **dictionary**.  
For each product, label the stock status as:
- `"OUT OF STOCK"` if stock is `0`
- `"LOW"` if stock is between `1` and `10` (inclusive)
- `"OK"` if stock is greater than `10`

---

### Input

```python
products = ["Laptop", "Mouse", "Keyboard", "Monitor", "Webcam", "Headset"]
stock    = [0, 5, 42, 3, 0, 17]
```

---

### Expected Output

```python
result = {
    'inventory_report': [
        {'index': 1, 'product': 'Laptop',   'qty': 0,  'status': 'OUT OF STOCK'},
        {'index': 2, 'product': 'Mouse',    'qty': 5,  'status': 'LOW'},
        {'index': 3, 'product': 'Keyboard', 'qty': 42, 'status': 'OK'},
        {'index': 4, 'product': 'Monitor',  'qty': 3,  'status': 'LOW'},
        {'index': 5, 'product': 'Webcam',   'qty': 0,  'status': 'OUT OF STOCK'},
        {'index': 6, 'product': 'Headset',  'qty': 17, 'status': 'OK'},
    ],
    'summary': {
        'total_products': 6,
        'out_of_stock': 2,
        'low': 2,
        'ok': 2
    }
}
```

---

### Constraints
- Must use `enumerate` (not `range(len(...))`) to generate the row number
- Row numbers are 1-based: `index` starts at 1
- Return a dict with exactly two keys: `'inventory_report'` (list of dicts) and `'summary'` (dict of counts)

````python
# Q1 — Your solution here
products = ["Laptop", "Mouse", "Keyboard", "Monitor", "Webcam", "Headset"]
stock    = [0, 5, 42, 3, 0, 17]

# out_of_stock, low, ok = 0, 0, 0
out_of_stock=low=ok=0
# products = [('Laptop', 0), ('Mouse', 5), ('Keyboard', 42)]
inventory_report = []
for idx, product in enumerate(zip(products, stock)):
    product, stock = product
    if stock==0:
        out_of_stock+=1
        status='Out of stock'
    elif stock>=1 and stock<=10:
        low+=1
        status='low'
    else:
        ok+=1
        status='Ok'
    
    product_detail = {
        'index': idx,
        'product': product,
        'quantity': stock,
        'status': status
    }
    inventory_report.append(product_detail)
summary = {
    'total_products': len(products),
    'out_of_stock': out_of_stock,
    'low': low,
    'ok': ok
}
result = {
    'inventory_report': inventory_report,
    'summary': summary
}
print(result)
````
